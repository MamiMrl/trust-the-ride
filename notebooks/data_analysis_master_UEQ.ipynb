```python
condition_A_df = df[df['condition'] == 'A']
condition_B_df = df[df['condition'] == 'B']
condition_C_df = df[df['condition'] == 'C']
```

```python
import plotly.express as px

# List of conditions
conditions = {
    'A': condition_A_df,
    'B': condition_B_df,
    'C': condition_C_df
}

# List of column names
columns = [
    'obstructive/supportive',
    'complicated/easy',
    'inefficient/efficient',
    'clear/confusing',
    'boring/exciting',
    'not interesting/interesting',
    'conventional/inventive',
    'usual/leading edge'
]

# Colors for each condition
colors = {
    'A': '#8BC1F7',
    'B': '#F9E0A2',
    'C': '#C9190B'
}

# Loop through each condition and each column
for condition, df in conditions.items():
    for column in columns:
        # Count the number of responses for each rating in the column
        rating_counts = df[column].value_counts().sort_index()

        # Create a bar chart for the condition and column
        fig = px.bar(
            rating_counts,
            x=rating_counts.index,
            y=rating_counts.values,
            labels={'x': 'Rating', 'y': 'Count'},
            title=f'{column.replace("_", " ").title()} Ratings for Condition {condition}',
            color_discrete_sequence=[colors[condition]]
        )

        fig.update_layout(
            plot_bgcolor='white',
            paper_bgcolor='white'
        )

        # Show the figure
        fig.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import shapiro, normaltest
import statsmodels.api as sm

# Define scales and conditions
scales = ['Pragmatic Quality', 'Hedonic Quality', 'Overall']
conditions = ['Functional UI (A)', 'No UI (B)', 'Placebo UI (C)']

# Function to plot histograms and Q-Q plots
def plot_normality(df, scale, condition):
    sns.histplot(df[scale], kde=True)
    plt.title(f'Histogram for {scale} under {condition}')
    plt.show()

    sm.qqplot(df[scale], line='s')
    plt.title(f'Q-Q plot for {scale} under {condition}')
    plt.show()

# Function to calculate descriptive statistics and perform normality tests
def check_normality(df, scale):
    descriptive_stats = df[scale].describe()
    skewness = df[scale].skew()
    kurtosis = df[scale].kurtosis()
    shapiro_test = shapiro(df[scale])
    normaltest_test = normaltest(df[scale])

    return {
        'descriptive_stats': descriptive_stats,
        'skewness': skewness,
        'kurtosis': kurtosis,
        'shapiro': shapiro_test,
        'normaltest': normaltest_test
    }

# Check normality for each scale within each condition
for condition in conditions:
    df_condition = combined_UEQ_df[combined_UEQ_df['condition'] == condition]
    for scale in scales:
        plot_normality(df_condition, scale, condition)
        stats = check_normality(df_condition, scale)
        print(f'Statistics for {scale} under {condition}:\n{stats}\n')
```

```python
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns

# Function to perform Levene's Test for homogeneity of variances
def levene_test(data, group_column, value_column):
    groups = data[group_column].unique()
    group_data = [data[data[group_column] == group][value_column] for group in groups]
    stat, p = stats.levene(*group_data)
    print(f"Levene's Test for {value_column} - Statistic: {stat}, p-value: {p}")

# Check homogeneity of variances for UEQ variables
ueq_columns = ['Pragmatic Quality', 'Hedonic Quality', 'Overall']
for column in ueq_columns:
    levene_test(combined_UEQ_df, 'condition', column)
```

```python
 subset_df = pd.DataFrame(columns=combined_UEQ_df.columns)
participant_ids = combined_UEQ_df['participant_id'].unique()

for pid in participant_ids:
    # Get rows for the current participant ID
    participant_data = combined_UEQ_df[combined_UEQ_df['participant_id'] == pid]

    # Check if the participant has more than one condition
    if participant_data['condition'].nunique() > 1:
        subset_df = pd.concat([subset_df, participant_data])

subset_df = subset_df.reset_index(drop=True)

print(subset_df)
```

```python
# Create subsets for Condition A-B and Condition A-C
condition_ab = subset_df[(subset_df['condition'] == 'Functional UI (A)') | (subset_df['condition'] == 'No UI (B)')]
condition_ac = subset_df[(subset_df['condition'] == 'Functional UI (A)') | (subset_df['condition'] == 'Placebo UI (C)')]

# Ensure that each participant has both conditions
condition_ab = condition_ab.groupby('participant_id').filter(lambda x: len(x) == 2)
condition_ac = condition_ac.groupby('participant_id').filter(lambda x: len(x) == 2)

# Reset index for convenience
condition_ab = condition_ab.reset_index(drop=True)
condition_ac = condition_ac.reset_index(drop=True)

# Display the subsets
print("\nSubset for Condition A-C:")
condition_ac
```
